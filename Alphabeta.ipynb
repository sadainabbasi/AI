**Alphabeta**

```python
import math

# Alpha-Beta Pruning function
def alpha_beta(depth, node_index, is_max, values, alpha, beta):

    # Terminal node condition
    if depth == 3:
        return values[node_index]

    # MAX player's turn
    if is_max:
        best = -math.inf

        for i in range(2):
            value = alpha_beta(
                depth + 1,
                node_index * 2 + i,
                False,
                values,
                alpha,
                beta
            )

            best = max(best, value)
            alpha = max(alpha, best)

            # Pruning condition
            if beta <= alpha:
                break

        return best

    # MIN player's turn
    else:
        best = math.inf

        for i in range(2):
            value = alpha_beta(
                depth + 1,
                node_index * 2 + i,
                True,
                values,
                alpha,
                beta
            )

            best = min(best, value)
            beta = min(beta, best)

            # Pruning condition
            if beta <= alpha:
                break

        return best


# Terminal values of the tree
values = [3, 5, 6, 9, 1, 2, 0, -1]

# Call Alpha-Beta Pruning
result = alpha_beta(
    0,              # depth
    0,              # root node index
    True,           # MAX starts
    values,
    -math.inf,      # alpha
    math.inf        # beta
)

print("Best value is:", result)
```

**HilClimb**

```python
# Hill Climbing Algorithm

# Function we want to maximize
def objective_function(x):
    return -(x - 3) ** 2 + 10


def hill_climbing(start):
    current = start

    while True:
        left = current - 1
        right = current + 1

        # Find best neighbor
        if objective_function(left) > objective_function(current):
            current = left

        elif objective_function(right) > objective_function(current):
            current = right

        else:
            # No better neighbor found
            return current


start_point = 0
best_solution = hill_climbing(start_point)

print("Best solution:", best_solution)
print("Maximum value:", objective_function(best_solution))
```

**LinearRegression**

```python
from sklearn.linear_model import LinearRegression

# Dummy data
x = [[2], [3], [4], [6]]
y = [4, 5, 6, 7]

# Create model
model = LinearRegression()

# Train model
model.fit(x, y)

# Predict value for x = 5
prediction = model.predict([[5]])

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
print("Prediction for x = 5:", prediction[0])
```
