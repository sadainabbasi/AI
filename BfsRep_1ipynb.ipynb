```python
graph={
     '5' : ['3','7'],
     '3' : ['2','4'],
     '7' : ['8'],
     '2' : [],
     '4' : ['8'],
     '8' :  [],
}
visited = []
queue = []
def bfs(visited,graph,node):
    visited.append(node)
    queue.append(node)

    while queue:
        m = queue.pop(0)
        print(m,end=" ")

        for neighbour in graph[m]: # Corrected indentation
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)

# Driver Code
print("following is the breadth-first search")
bfs(visited,graph,'5') # Corrected syntax and moved outside function
```
