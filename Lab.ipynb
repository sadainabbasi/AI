**Lab Task 2: Object-Oriented Design and Area Comparison**

```python
"""
1. Design two classes representing Trapezoid and Parallelogram.
Define appropriate attributes and compute their areas using standard mathematical formulas.
"""

class Trapezoid:
    def __init__(self, a, b, h):
        self.a = a
        self.b = b
        self.h = h

    def area(self):
        return 0.5 * (self.a + self.b) * self.h


class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height



# 2. Develop a comparison mechanism to evaluate areas of different shapes.

t = Trapezoid(3, 5, 4)
p = Parallelogram(6, 4)

if t.area() > p.area():
    print("Trapezoid is bigger")
else:
    print("Parallelogram is bigger")



# 3. Store the computed areas in a dictionary structure.
areas = {
    "Trapezoid": t.area(),
    "Parallelogram": p.area()
}
print(areas)


# 4. Iterate over multiple sets of dimensions to analyze variations in area.

for i in range(3):
    t = Trapezoid(i+2, i+3, i+4)
    print("Area:", t.area())



#5. Utilize dictionary operations to determine and display the shape with the maximum area.

areas = {
    "T1": Trapezoid(3, 5, 4).area(),
    "P1": Parallelogram(6, 4).area()
}
max_shape = max(areas, key=areas.get)
print("Max area:", max_shape, areas[max_shape])





```
