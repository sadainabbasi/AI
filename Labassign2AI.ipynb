**Constraint Satisfaction Problem**

```python
# Question 1: Constraint Satisfaction Problem (CSP)
# Assigning courses (AI, DS, CN) to classrooms (Room A, Room B, Room C)
# using the Backtracking Search Algorithm.

# Step 1: Define Variables and Domains
# ------------------------------------
variables = ["AI", "DS", "CN"]

# Every course can initially be placed in any room
domains = {
    "AI": ["Room A", "Room B", "Room C"],
    "DS": ["Room A", "Room B", "Room C"],
    "CN": ["Room A", "Room B", "Room C"],
}

# Step 2: Define Constraints
# ---------------------------------------------------------
def is_consistent(variable, value, assignment):
    """
    Checks whether assigning 'value' (a room) to 'variable' (a course)
    is consistent with the constraints, given the current partial assignment.
    """

    # Constraint 1: AI cannot be assigned to Room A
    if variable == "AI" and value == "Room A":
        return False

    # Constraint 2: DS cannot be assigned to Room C
    if variable == "DS" and value == "Room C":
        return False

    # Constraint 4: CN must be Room A or Room C
    if variable == "CN" and value not in ["Room A", "Room C"]:
        return False

    # Constraint 3: No two courses can share the same classroom
    for assigned_var, assigned_val in assignment.items():
        if assigned_val == value:
            return False

    return True


# ---------------------------------------------------------
# Step 3: Backtracking Search Algorithm
# ---------------------------------------------------------
def backtracking_search(assignment, variables, domains):
    # Base case: all variables assigned -> solution found
    if len(assignment) == len(variables):
        return assignment

    # Pick the next unassigned variable
    unassigned = [v for v in variables if v not in assignment]
    current_var = unassigned[0]

    # Try each value in the domain of the current variable
    for value in domains[current_var]:
        if is_consistent(current_var, value, assignment):
            assignment[current_var] = value          # assign
            result = backtracking_search(assignment, variables, domains)
            if result is not None:
                return result
            del assignment[current_var]               # backtrack

    return None  # no valid value found -> trigger backtracking


# ---------------------------------------------------------
# Step 4: Run and Display Result
# ---------------------------------------------------------
if __name__ == "__main__":
    solution = backtracking_search({}, variables, domains)

    print("---- CSP: Classroom Assignment ----")
    if solution:
        for course, room in solution.items():
            print(f"{course} -> {room}")
    else:
        print("No valid assignment found.")
```

**Grading System**

```python
# Question 2: Knowledge-Based System (KBS)
# Recommends a student's grade based on marks, using an explicit rule base.

# Step 1: Represent the Knowledge Base

# Each rule = (condition function, conclusion)
knowledge_base = [
    (lambda marks: marks >= 85, "A"),
    (lambda marks: 70 <= marks < 85, "B"),
    (lambda marks: 50 <= marks < 70, "C"),
    (lambda marks: marks < 50, "Fail"),
]

# ---------------------------------------------------------
# Step 2: Inference Engine
# ---------------------------------------------------------
def infer_grade(marks):
    """
    Applies the rules in order (forward chaining style) and
    returns the first matching conclusion.
    """
    for condition, conclusion in knowledge_base:
        if condition(marks):
            return conclusion
    return "Invalid marks"  # fallback, e.g. negative input


# ---------------------------------------------------------
# Step 3: Take Input and Display Result
# ---------------------------------------------------------
def get_grade_for(marks):
    grade = infer_grade(marks)
    print(f"Marks: {marks} -> Grade: {grade}")


if __name__ == "__main__":
    print("---- Knowledge-Based System: Student Grading ----")

    # Test with at least three different inputs (as required)
    test_marks = [90, 75, 60, 40]
    for m in test_marks:
        get_grade_for(m)

    # Optional: allow interactive input too
    try:
        user_marks = int(input("\nEnter marks to check your grade: "))
        get_grade_for(user_marks)
    except ValueError:
        print("Please enter a valid number next time.")
```
