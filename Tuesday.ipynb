```python
# Python program to print DFS traversal for complete graph
from collections import defaultdict
# This class represents a directed graph using adjacency
# list representation
class Graph:
    # Constructor
    def __init__(self):
        # default dictionary to store graph
        self.graph = defaultdict(list)

    # function to add an edge to graph
    def addEdge(self,u,v):
        self.graph[u].append(v)

    # A function used by DFS
    def DFSUtil(self, v, visited):
        # Mark the current node as visited and print it
        visited[v]= True
        print(v, end=' ')

        # Recur for all the vertices adjacent to
        # this vertex
        for i in self.graph[v]:
            if visited[i] == False:
                self.DFSUtil(i, visited)

    # The function to do DFS traversal. It uses
    # recursive DFSUtil()
    def DFS(self):
        # Determine the maximum node index to initialize visited array correctly
        # Collect all unique nodes from keys and values to find the actual number of vertices
        all_nodes = set(self.graph.keys())
        for key in self.graph:
            all_nodes.update(self.graph[key])

        V = 0
        if all_nodes:
            V = max(all_nodes) + 1 # V represents the size of the visited array needed for 0-indexed nodes

        # Mark all the vertices as not visited
        visited =[False]*(V)

        # Call the recursive helper function to print
        # DFS traversal starting from all vertices one
        # by one, ensuring all components are covered
        for i in range(V):
            # Ensure that we only attempt to start DFS from a node that actually exists in the graph
            if i in all_nodes and visited[i] == False:
                self.DFSUtil(i, visited)

# Driver code
# Create a graph given in the above diagram
g = Graph()
g.addEdge(0, 1)
g.addEdge(0, 2)
g.addEdge(1, 2)
g.addEdge(2, 0)
g.addEdge(2, 3)
g.addEdge(3, 3)
print("Following is Depth First Traversal")
g.DFS()
```

```python
from collections import deque

class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v):
        if u not in self.graph:
            self.graph[u] = []
        if v not in self.graph:
            self.graph[v] = []
        self.graph[u].append(v)
        self.graph[v].append(u)

    def bfs(self, start, goal):
        visited = set()
        queue = deque([start])
        visited.add(start)
        count = 0

        while queue:
            node = queue.popleft()
            count += 1
            print(node, end=" ")

            if node == goal:
                print("\nGoal found!")
                return count

            for neighbor in self.graph[node]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor)

        return count

    def dfs(self, start, goal):
        visited = set()
        stack = [start]
        count = 0

        while stack:
            node = stack.pop()
            if node not in visited:
                visited.add(node)
                count += 1
                print(node, end=" ")

                if node == goal:
                    print("\nGoal found!")
                    return count

                # reverse for consistent order
                for neighbor in reversed(self.graph[node]):
                    if neighbor not in visited:
                        stack.append(neighbor)

        return count


# 🔷 Create Graph
g = Graph()
g.add_edge('A', 'B')
g.add_edge('A', 'C')
g.add_edge('A', 'D')
g.add_edge('B', 'E')
g.add_edge('C', 'E')
g.add_edge('D', 'E')

# 🔷 Case 1: BFS < DFS
print("Case 1: Goal = C")
print("BFS Traversal:")
bfs_count = g.bfs('A', 'C')

print("DFS Traversal:")
dfs_count = g.dfs('A', 'C')

print("Nodes visited in BFS:", bfs_count)
print("Nodes visited in DFS:", dfs_count)

print("\n----------------------\n")

# 🔷 Case 2: BFS > DFS
print("Case 2: Goal = E")
print("BFS Traversal:")
bfs_count = g.bfs('A', 'E')

print("DFS Traversal:")
dfs_count = g.dfs('A', 'E')

print("Nodes visited in BFS:", bfs_count)
print("Nodes visited in DFS:", dfs_count)
```

```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v, cost):
        if u not in self.graph:
            self.graph[u] = []
        if v not in self.graph:
            self.graph[v] = []

        self.graph[u].append((v, cost))
        self.graph[v].append((u, cost))  # undirected graph

    def dfs(self, start, goal):
        stack = [(start, [start], 0)]  # (node, path, distance)
        visited = set()

        while stack:
            node, path, dist = stack.pop()

            if node not in visited:
                visited.add(node)

                if node == goal:
                    return path, dist

                for neighbor, cost in reversed(self.graph[node]):
                    if neighbor not in visited:
                        stack.append((neighbor, path + [neighbor], dist + cost))

        return None, 0


# 🔷 Create Graph (Romania Map)
g = Graph()

g.add_edge("Arad", "Zerind", 75)
g.add_edge("Arad", "Sibiu", 140)
g.add_edge("Arad", "Timisoara", 118)

g.add_edge("Zerind", "Oradea", 71)
g.add_edge("Oradea", "Sibiu", 151)

g.add_edge("Sibiu", "Fagaras", 99)
g.add_edge("Sibiu", "Rimnicu Vilcea", 80)

g.add_edge("Fagaras", "Bucharest", 211)

g.add_edge("Rimnicu Vilcea", "Pitesti", 97)
g.add_edge("Pitesti", "Bucharest", 101)

g.add_edge("Timisoara", "Lugoj", 111)
g.add_edge("Lugoj", "Mehadia", 70)
g.add_edge("Mehadia", "Drobeta", 75)
g.add_edge("Drobeta", "Craiova", 120)
g.add_edge("Craiova", "Pitesti", 138)


# 🔷 Run DFS
start = "Arad"
goal = "Bucharest"

path, distance = g.dfs(start, goal)

print("DFS Path:", " → ".join(path))
print("Total Distance:", distance)
```
