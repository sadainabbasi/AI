```python
class Node:
    def __init__(self, value=None, children=None):
        self.value = value
        self.children = children if children else []

def minimax(node, is_max):
    # Base case: leaf node
    if not node.children:
        return node.value

    if is_max:
        best = float('-inf')
        for child in node.children:
            val = minimax(child, False)
            best = max(best, val)
        return best
    else:
        best = float('inf')
        for child in node.children:
            val = minimax(child, True)
            best = min(best, val)
        return best


# Leaf nodes
S = Node(-4)
T = Node(-5)
M_leaf = Node(2)
N_leaf = Node(3)
R_leaf = Node(4)
Q = Node(5)
A = Node(6)
B = Node(7)

# Level 2
D = Node(children=[S, T])         # max
R = Node(children=[M_leaf, N_leaf])  # max
O = Node(children=[R_leaf, Q])    # max
P = Node(children=[A, B])         # max

# Level 1
M = Node(children=[D, R])         # min
N = Node(children=[O, P])         # min

# Root
Z = Node(children=[M, N])         # max

# Run minimax
result = minimax(Z, True)
print("Optimal value:", result)
```

```python
class Node:
    def __init__(self, value=None, children=None):
        self.value = value
        self.children = children if children else []

def alpha_beta(node, alpha, beta, is_max):
    # Leaf node
    if not node.children:
        return node.value

    if is_max:
        value = float('-inf')
        for child in node.children:
            value = max(value, alpha_beta(child, alpha, beta, False))
            alpha = max(alpha, value)

            if alpha >= beta:  # Pruning
                break
        return value
    else:
        value = float('inf')
        for child in node.children:
            value = min(value, alpha_beta(child, alpha, beta, True))
            beta = min(beta, value)

            if beta <= alpha:  # Pruning
                break
        return value


# Leaf nodes
n4 = Node(4)
n3 = Node(3)
n8 = Node(8)
n9 = Node(9)
n0 = Node(0)
n1 = Node(1)
n8b = Node(8)
n5 = Node(5)

# Level 2 (MAX nodes)
D = Node(children=[n4, n3])
E = Node(children=[n8, n9])
F = Node(children=[n0, n1])
G = Node(children=[n8b, n5])

# Level 1 (MIN nodes)
B = Node(children=[D, E])
C = Node(children=[F, G])

# Root (MAX)
A = Node(children=[B, C])

# Run algorithm
result = alpha_beta(A, float('-inf'), float('inf'), True)
print("Optimal value:", result)
```
